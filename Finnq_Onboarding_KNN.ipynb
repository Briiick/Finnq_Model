```
# IMPORT LIBRARIES
import pandas as pd
import numpy as np
import math
import statistics as stat
import random
from scipy.stats import skewnorm
import pickle
from sklearn.model_selection import train_test_split, cross_validate
from sklearn import preprocessing, neighbors
from sklearn.neighbors import KNeighborsClassifier
from sklearn import metrics
from sklearn.metrics import classification_report, confusion_matrix
import matplotlib.pyplot as plt
```

```
#KNN model on demographic data 
df = pd.read_csv('supervised_4.csv')

#data processing: convert monthly incomes to ints, 
for i in range(len(df['monthly_inc'])):
    df['monthly_inc'][i] = df['monthly_inc'][i].replace(",", "")
    df['monthly_inc'][i] = int(df['monthly_inc'][i])
    
    
#create new dataframe without labels
df_labeless = df.drop(['persona', 'age'], 1)
df_labeless['default'] = df_labeless['default'] * 1000

# create our X predictors and Y label
X = np.array(df_labeless)
y = np.array(df['persona'])

#splitting data into training and test sets (70:30)
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=4)

```

```
#plot results
plt.style.use('fivethirtyeight')
plt.figure(num=None, figsize=(15, 15))
plt.plot(k_range, scores_list)
plt.xlabel('Value of K for KNN')
plt.ylabel('Testing Accuracy')
```
